```python
"""
To hashed lists that contain dictionaries or identification numbers
The hash function was used for anonymization as hashing keys is part of anonymization operations,
 comparing data without revealing values to check if two data sets are similar without revealing the real data.
The SHA-256 algorithm was used to achieve specific privacy requirements.
"""
```

```python

import hashlib

def hash_list(id_list):
    concatenated_ids = ''.join(map(str, id_list))
    hashed_ids = hashlib.sha256(concatenated_ids.encode('utf-8')).hexdigest()

    return hashed_ids

def hash_dict_keys(dictionary):
    sorted_keys = sorted(dictionary.keys())
    return hash_list(sorted_keys)

def are_dicts_equal(dict_1, dict_2):
    hdk= hash_dict_keys(dict_1) == hash_dict_keys(dict_2) and dict_1 == dict_2
    return hdk

# Example usage:
name = {'ahmad': 1, 'mohammad': 2, 'mahmoud': 3}
id = {'1598': 5, '1588': 6, '1533': 7}
id_2 = {'1588': 6, '1533': 7, '1598': 5}

are_equal1 = are_dicts_equal(name, id_2)
are_equal2 = are_dicts_equal(id_2, id)

print("Dictionaries 1 and 2 are equal:", are_equal1)
print("Dictionaries 1 and 3 are equal:", are_equal2)
Sura
Sura Musa
communicate private.py
Sura
Sura Musa
"""To encrypt conversations between any two parties to prevent any third party
from accessing or guessing them, and ultimately to reveal by the  the content of the message,
which may contain sensitive data that would worry peopleâ€™s comfort if it were accessed,
 as it is revealed through the decryption key."""

print("please enter your message: ")

def encrypt_msg(message, key):
    encrypted_message = ''.join(chr(ord(c) ^ key) for c in message)
    return encrypted_message

def decrypt_msg(encrypted_message, key):
    decrypted_message = ''.join(chr(ord(c) ^ key) for c in encrypted_message)
    return decrypted_message



# Example usage:
encrypt_key = 8
message = "Hello, I Live in Egypt"
encrypted_message = encrypt_msg(message ,encrypt_key)
decrypt_key = 8
decrypted_message = decrypt_msg(encrypted_message, decrypt_key)
def check_msg():
  if decrypted_message == encrypt_msg(encrypted_message , 8):
    print(message)
  else:
    print("not same msg")

check_msg()
```
