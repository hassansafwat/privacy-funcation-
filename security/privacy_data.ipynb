```python
"""The provided code includes several techniques related to privacy and data protection.
 It provides a function that restricts access to data based on user roles to ensure that users only
  access data relevant to their role. The "anonymize_data" function works to anonymize user data by
  replacing specific attributes ( Such as names) with generic values (such as “anonymous”).
  The mask_ssn function also hides Social Security Numbers (SSNs) by replacing them with asterisks.
The hash_password function hashes user passwords using the SHA-256 algorithm."""
```

```python

import hashlib
# Sample dataset
data = [
    {'name': 'Alice', 'ssn': '123-45-6789', 'email': 'alice@example.com'},
    {'name': 'Bob', 'ssn': '987-65-4321', 'email': 'bob@example.com'},
    # ... more data
]
# Access Control (Simple Example)
class User:
    def __init__(self, role):
        self.role = role
def check_access(user, data):
    if user.role == 'admin':
        return data
    elif user.role == 'user':
        return [{'name': entry['name']} for entry in data]
    else:
        return []
# Data Anonymization
def anonymize_data(data):
    andata= [{'name': 'Anonymous'} for _ in data]
    return andata

# Data Masking (Masking SSNs)
def mask_ssn(data):
    for entry in data:
        if 'ssn' in entry:
            entry['ssn'] = '*-**-**'

# Hashing Passwords
def hash_password(password):
    hpass = hashlib.sha256(password.encode()).hexdigest()
    return hpass

# Usage Example
if _name_ == '_main_':
    # Simulate user roles
    admin_user = User(role='admin')
    user_user = User(role='user')

    # Access control
    admin_data = check_access(admin_user, data)
    user_data = check_access(user_user, data)

    # Anonymize user data
    anonymized_data = anonymize_data(data)

    # Mask SSNs
    masked_data = data.copy()
    mask_ssn(masked_data)

    # Hash passwords
    password = 'my_secure_password'
    hashed_password = hash_password(password)

    print("Admin's Data:")
    print(admin_data)

    print("\nUser's Data:")
    print(user_data)

    print("\nAnonymized Data:")
    print(anonymized_data)

    print("\nMasked Data:")
    print(masked_data)

    print(f"\nHashed Password: {hashed_password}")
```
